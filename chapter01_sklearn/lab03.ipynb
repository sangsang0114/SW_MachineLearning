```python
import mglearn
import pandas as pd 
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import matplotlib
from matplotlib import font_manager, rc
```

```python
font_location = "c:/Windows/fonts/malgun.ttf"
font_name = font_manager.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family=font_name)
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
# 군집 모델을 만듭니다
kmeans = KMeans(n_clusters=3)
kmeans.fit(iris.iloc[:,0:4])
# print("클러스터 레이블 ", kmeans.labels_) 
```

```python
def show_plot(x,y):
    mglearn.discrete_scatter(iris[x],iris[y], kmeans.labels_)
    plt.legend(["클러스터 0", "클러스터 1", "클러스터 2"], loc='best')
    plt.xlabel(x)
    plt.ylabel(y)
    plt.show()
```

```python
def show_plot2(x,y):
    import seaborn as sns
    sns.scatterplot(x = x, y= y, hue = 'Name', data = iris)
```

다음 그래프를 생성하세요. 1-1) SepalLength,SepalWidth 을 각각 x,y 축 데이터로 산점도 그래프를 그리세요. ( 클러스터 LABEL 별 다른 색으로 표현)

```python
show_plot('SepalLength', 'SepalWidth')
```

다음 그래프를 생성하세요. 1-2) SepalLength,SepalWidth 을 각각 x,y 축 데이터로 산점도 그래프를 그리세요. ( NAME 속성 값 별로 다른 색으로 표현)
