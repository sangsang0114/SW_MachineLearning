```python
import numpy as np
import matplotlib.pyplot as plt
import ex_RNN_2_timeSeries_data as ex_t2_d
import ex_RNN_2_timeSeries_model as ex_t2_m

## load Data  
(X_train2,Y_train2) =ex_t2_d.make_data_Sequence2Sequence()

## make Model 
model2= ex_t2_m.make_model() 
```

### 학습 전 예측

```python
plt.subplot(211) # 2행 1열 1번째 그래프 
plt.plot([1, 2, 3], X_train2[1].flatten(), 'bo-', label="input sequence")
plt.plot([4, 5, 6], Y_train2[1].flatten(), 'ro-', label="target sequence")
plt.plot([4, 5, 6], model2.predict(X_train2[1:2,:,:]).flatten(), 'gs-', label="output sequence")
plt.xlim(-0.5, 7.5)
plt.ylim(-1.1, 1.1)
plt.legend()
plt.title("Second sample sequence")


plt.subplot(212) # 2행 1열 2번째 그래프 
plt.plot([2, 3, 4], X_train2[1].flatten(), 'bo-', label="input sequence")
plt.plot([5, 6, 7], Y_train2[1].flatten(), 'ro-', label="target sequence")
plt.plot([5, 6, 7], model2.predict(X_train2[2:3,:,:]).flatten(), 'gs-', label="output sequence")
plt.xlim(-0.5, 7.5)
plt.ylim(-1.1, 1.1)
plt.legend()
plt.title("Third sample sequence")
plt.tight_layout()
plt.show()

# 초록색 선 : 예측 값.  빨간색 선 : target 값  
# 학습 전이기 때문에  예측 값 이 target 값과 많은 차이가 있음.
```

```python
history2 = model2.fit(X_train2, Y_train2, epochs=1000, verbose=0)
plt.plot(history2.history["loss"])
plt.title("Loss")
plt.show()
```
